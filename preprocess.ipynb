```python
#Import library
import pandas as pd
import numpy as np
import os
```

```python
def raw_data(file_sed):
    """
    - This function will convert the raw file sed in folder Spectral reflectance measurement into a numpy array
    - parameter:
      file_sed: takes the file names in a specific format
    - return:
      df_sed_file: return a dataframe of each sed file which contains Wavelength and Reflection
    """
    list_data = [] #empty list to store the split data
    f = open(file_sed,"r")    

    arr = f.readlines() #Read all the line in the sed file         
    index = arr.index("Data:\n") + 1 #Start reading the file from Row Data  
    arr1 = arr[index:len(arr)] #Read the data from row Data to the end

    for i in arr1: 
        var = i.split("\t") # Separate the info in each collumn
        list_data.append(var) 

    df_sed_file = pd.DataFrame(np.array(list_data[1:len(list_data)]), columns=list_data[0])

    df_sed_file[['Reflect. %\n']] = df_sed_file[['Reflect. %\n']].astype(float)

    return df_sed_file[['Reflect. %\n']]
```

```python
def through_file():
    """
    Read each file in the folder Spectral Reflectance measurement
    """
    directory = 'C:\\Users\\Admin\\OneDrive\\Tài liệu\\Spectral reflectance measurement'
    list_result = []
    # iterate over files in that directory
    for filename in os.listdir(directory):
        f = os.path.join(directory, filename)
    # checking if it is a file
        if os.path.isfile(f):
            result = raw_data(f)
            list_result.append(result)

    return list_result
```

```python
def process_filename():
    """
    Split file name of the file in the folder 
    """
    directory = 'C:\\Users\\Admin\\OneDrive\\Tài liệu\\Spectral reflectance measurement'
    list_result = []
    for filename in os.listdir(directory):
        f = os.path.join(directory, filename)
        f = str(f) 
        # checking if it is a file
        if os.path.isfile(f):
            a = f.split("_")
            a = a[2].rstrip('.sed') #remove the .sed 
            list_result.append(a)
    return list_result
```

```python
def process_csv():
    """ 
    Processing data in the file Data-mua1.csv
    return: a Dataframe with their replicate and P, K, Chlorophyll-a
    """
    df = pd.read_csv("C:\\Users\\Admin\\OneDrive\\Tài liệu\\GroupProject-main\\GroupProject-main\\DATA_Mua1_2022.csv")
    df = df[df["Sub-Replicate"].notnull()] #Create new dataframe based on sub-replicate
    df.loc[:, ['Replicate']] = df.loc[:,['Replicate']].ffill() #Input the misssing replicate in data
    df["Sub-Replicate"] = df["Sub-Replicate"].astype(int)
    df["Sub-Replicate"] = df["Sub-Replicate"].astype(str)
    #new_df = new_df.tail(10)
    df['Replicate and Sub'] = df['Replicate'] + "." + df['Sub-Replicate']
    for i in range(1,7):
        df['Replicate and Sub'].replace(to_replace = f"BC{i}.1", value = f"BC{i}", regex=True, inplace=True)
    
    new_df = df[['Replicate and Sub', 'P conc. (mg/kg)', 'K conc. (mg/kg)', 'Chlorophyll-a']]

    return new_df
```

```python
DATA_Mua1_2022 = process_csv()
```

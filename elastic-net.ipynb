```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
from sklearn.linear_model import ElasticNet
```

```python
df = pd.read_csv('Output-database.csv')
print(df.head(10))
```

```python
scaler = StandardScaler()
```

```python
X = df.iloc[: , 0:2151]
Y = df['Chlorophyll-a']
```

```python
Y1 = df['P conc. (mg/kg)']
```

```python
Y2 = df['K conc. (mg/kg)']
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, Y, test_size=0.4, random_state=101)
```

```python
X_train, X_test, y1_train, y1_test = train_test_split(
    X, Y1, test_size=0.4, random_state=101)
```

```python
X_train, X_test, y2_train, y2_test = train_test_split(
    X, Y2, test_size=0.4, random_state=101)
```

```python
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.fit_transform(X_test)
```

```python
from sklearn.model_selection import GridSearchCV

# Create the random grid
random_grid = {"max_iter": [1, 2, 5, 10, 20, 30, 40, 50, 60, 70, 80, 100],
                "alpha": [0.0001, 0.001, 0.01, 0.1, 1, 10, 100],
                "l1_ratio": np.arange(0.0, 1.0, 0.1)}
grid_ridge_performance = GridSearchCV(ElasticNet(), param_grid = random_grid, scoring = 'r2')
grid_ridge_performance.fit(X_train_scaled, y_train)
grid_ridge_performance.best_params_
```

```python
from sklearn.model_selection import GridSearchCV

# Create the random grid
random_grid = {"max_iter": [1, 2, 5, 10, 20, 30, 40, 50, 60, 70, 80, 100],
                "alpha": [0.0001, 0.001, 0.01, 0.1, 1, 10, 100],
                "l1_ratio": np.arange(0.0, 1.0, 0.1)}
grid_ridge_performance = GridSearchCV(ElasticNet(), param_grid = random_grid, scoring = 'r2')
grid_ridge_performance.fit(X_train_scaled, y2_train)
grid_ridge_performance.best_params_
```

```python
model = ElasticNet(alpha= 10, l1_ratio= 0.0, max_iter= 20)
model.fit(X_train_scaled,y_train)

model2 = ElasticNet(alpha= 10, l1_ratio= 0.7000000000000001, max_iter= 10)
model2.fit(X_train_scaled,y1_train)

model3 = ElasticNet(alpha= 100, l1_ratio= 0.8, max_iter= 40)
model3.fit(X_train_scaled,y2_train)
```

```python
predictions = model.predict(X_test_scaled)
print(predictions)
```

```python
predictions1 = model2.predict(X_test_scaled)
print(predictions1)
```

```python
predictions2 = model3.predict(X_test_scaled)
print(predictions2)
```

```python
print(
  'R square of Chlorophyll-a: ', r2_score(y_test, predictions),
)
print(
  'Mean squared error of Chlorophyll-a: ', mean_squared_error(y_test, predictions),
)
```

```python
print(
  'R square of P: ', r2_score(y1_test, predictions1),
)
print(
  'Mean squared error of P: ', mean_squared_error(y1_test, predictions1),
)
```

```python
print(
  'R square of K: ', r2_score(y2_test, predictions2),
)
print(
  'Mean squared error of K: ', mean_squared_error(y2_test, predictions2),
)
```

```python
def mape(y_test, pred):
    y_test, pred = np.array(y_test), np.array(pred)
    mape = np.mean(np.abs((y_test - pred) / y_test))
    return mape
```

```python
print('Mean absolute percentage eror of Chlorophyll-a: ', mape(y_test, predictions))
```

```python
print('Mean absolute percentage eror of P: ', mape(y1_test, predictions1))
```

```python
print('Mean absolute percentage eror of K: ', mape(y2_test, predictions2))
```
